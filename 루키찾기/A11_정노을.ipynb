```python
#전체 crawling

driver = webdriver.Chrome() 
url = 'https://product.kyobobook.co.kr/bestseller/total?period=003#?page=1&per=20&period=003&ymw=&bsslBksClstCode=A'
driver.get(url)
#배너 끄기
driver.find_element(By.XPATH, '/html/body/div[18]/button').click()
driver.find_element(By.XPATH, '//*[@id="ui-id-66"]/div[2]/button[2]/span').click()


kb_df = pd.DataFrame()

def find_text_list(url):
    text_list = driver.find_elements(By.XPATH, url)
    return [i.text for i in text_list]

def push(url):
    #월간 버튼 누르기
    driver.find_element(By.XPATH, '//*[@id="selListType-button"]/span[1]').click()
    #기간 누르기
    driver.find_element(By.XPATH, url).click()

for i in range(12):   #기간 갯수, range범위 12으로 바꾸기
    #스크롤 올리기
    driver.execute_script("window.scrollTo(0, -document.body.scrollHeight);")
    time.sleep(2)
    #월간 버튼누르기
    push('//*[@id="selListType-menu"]/li[' + str(i + 1) + ']') 
    time.sleep(2)
    
    for j in tqdm(range(3)):   #페이지 갯수 10개로 바꾸기
        time.sleep(5)
        title = find_text_list('//*[@id="tabRoot"]/div[3]/ol/li/div[2]/div[2]/div[3]/div/div/a')
        author_etc = find_text_list('//*[@id="tabRoot"]/div[3]/ol/li/div[2]/div[2]/span')
        review_count = find_text_list('//*[@id="tabRoot"]/div[3]/ol/li/div[2]/div[2]/div[5]/div/span[1]/span[2]')

        result = pd.DataFrame()

        result['Title'] = title
        result['author_etc'] = author_etc
        result['review_count'] = review_count

        #ignore_index = True으로 기존 인덱스 무시하고 새롭게 설정
        kb_df = pd.concat([kb_df, result], ignore_index = True)

        time.sleep(3)
        #스크롤 내리기
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        #페이지 넘기기
        driver.find_element(By.XPATH, '//*[@id="tabRoot"]/div[3]/div[2]/button[2]').click()
       
```

```python
#author_etc 칼럼 고치기
#분해하고 싶은 칼럼 리스트에 넣음
auth_list = list(kb_df['author_etc'])

#칼럼명으로 들어갈 리스트 새성
Author = []
Publisher = []
Date_of_publication = []

#하나의 요소를 txt 변수에 넣고,  · 기준으로 split
for txt in auth_list:
    split_data = txt.split(' · ')
    Author.append(split_data[0])
    Publisher.append(split_data[1])
    Date_of_publication.append(split_data[2])
    
#칼럼 생성해서 각 리스트 넣기
kb_df['Author'] = Author
kb_df['Publisher'] = Publisher
kb_df['Date_of_publication'] = Date_of_publication

#필요없는 author_etc 지우기
kb_df.drop(columns=['author_etc'], inplace=True)

#review_count 칼럼 고치기

review = list(kb_df['review_count'])

Review_count = []
p = '(.*)개의 리뷰'

for i in review:
    #re.search() : 첫번째로 찾은 패턴의 양쪽 끝 인덱스 반환
    #정규표현식을 사용해서 개수만 추출
    num = re.search(p, i).group(1)
    #'(' 이거 없애기
    num = num.replace('(', '')
    Review_count.append(num)
    
kb_df['Review_count'] = Review_count

#필요없는 review_count 지우기
kb_df.drop(columns=['review_count'], inplace=True)
```
